<a id='mechanism'></a>
Code a mechanism that is able to detect exactly the identified failure types automatically. This could then be used for autodetecting the known failures in other datasets. The output should be a list including:
- The time of failure
- The failure type (of the failures you previously identified)

```python
def detect_failure(df_pivot):
    """
    Identify failure in sensor reading data mechanism based on predefined conditions.
    These conditions were extracted after being tested on previous sensor data.
    
    Attributes: 
        df (pd.DataFrame): all sensor readings in pivot format
                    
    Return:
        failure_dict (dict) : It has failure type as key and list of corresponding timestamps as value
        
    """
    # Define failure type dict with key as type and condition as value
    failure_types ={
    'vehicle_speed':((df_pivot['vehicle_speed']==0)&(df_pivot['transmission_state']=='drive')),
    'transmission':((df_pivot['vehicle_speed']>0)&(df_pivot['transmission_state']!='drive')),
    'engine_state':((df_pivot['vehicle_speed']>0)&(df_pivot['engine_state']=='engine_off')),
    'road_type':((df_pivot['road_type']=='city') & ~(df_pivot['vehicle_speed'] <= 60)) | \
((df_pivot['road_type']=='country_road') & \
         ~ ((df_pivot['vehicle_speed'] >60) & (df_pivot['vehicle_speed'] <=110))) |\
((df_pivot['road_type']=='highway') & ~(df_pivot['vehicle_speed'] >110))
    }
    
    # Extract failure timestamps for predefined failure types.
    failure_time_dict = {}
    
    for failure,condition in failure_types.items():
        # Check if there exists timestamps for related failure or not
        if len(list(df_pivot[condition].index)) > 0:
            failure_time_dict[failure] = list(df_pivot[condition].index)
        
    return  failure_time_dict
```

```python
failure_dict = detect_failure(df_pivot)
```

### Total time driven

```python
# Convert the data type of timestamp column to datatime format
warnings.filterwarnings("ignore")

def calc_time(df):
    df['date'] = pd.to_datetime(df.index)
    
    time_driven = df.date.max() - df.date.min()
    out_dict = {}
    
    out_dict['days'] = time_driven.days
    
    hours, remainder = divmod(time_driven.seconds, 3600)
    total_time_in_hrs = time_driven.seconds/3600
    
    out_dict['hours'] = hours
    minutes, seconds = divmod(remainder, 60)
    out_dict['minutes'] = minutes
    out_dict['seconds'] = seconds+ time_driven.microseconds / 1e6
    
    return out_dict, total_time_in_hrs

# Filter by engine_on and in drive mode
driven_data = df_pivot[(df_pivot['engine_state']=='engine_on')&(df_pivot['transmission_state']=='drive')]
total_time_driven,hrs_driven = calc_time(driven_data)
print(f'Total time dirven: {hrs_driven:.{3}f} hour(s).')
```

### Failure statistics.

In order to calculate time & distance driven during failure is considered challenging task because for each failure type we can not substract max. timestamp by minimum the reason why is because between max. timestamp and minimum there exist some correct readings so we need to cut the dataframe into failure intervals where each interval has a start timestamp and end one. [Diagram](#all_failures)

**Algorithm** used: 
<img src="detect_intervals.png">

```python
def get_failure_intervals(failure_df, failure, timestamps):
    """
    Extract failure intervals from df. This done by cutting all data into intervals.
    
    Those intervals are defined by start and end.
    Start intervals refers to the exact timestamp that sensor starts to fail in select failure type.
    End interval refers to the exact timestamp that sensor returns to work correctly.
    
    
    Attributes: 
        failure_df (pd.DataFrame): All sensor readings in pivot format
        failure (str): Failure Type name
        timestamps (list): List of all timestamps that have failure reading. 
                    
    Return:
        intervals (list) : List of extracted failure intervals.      
    """
    failure_df['timestamp'] = failure_df.index
    start_intervals = []
    end_intervals = []

    
    failure_name = failure+'_failure'
    failure_df.loc[timestamps,failure_name]=1 
    failure_df[failure_name].fillna(0, inplace=True)
    # Get the next state for next timestamp
    failure_df["next"] = failure_df[failure_name].shift(-1)
 
    start_intervals = list(failure_df[((failure_df[failure_name]==0)&(failure_df['next']==1))]['timestamp'])
    end_intervals = list(failure_df[((failure_df[failure_name]==1)&(failure_df['next']==0))]['timestamp'])

    # Check if the sensor from begining was not working well
    # If that is the case:
    # Add the first time stamp as start_interval   
    if len(end_intervals) > len(start_intervals) :
        start_intervals = [failure_df.timestamp.min()] + start_intervals
    
    intervals = list(zip(start_intervals, end_intervals))
    return intervals
```

```python
# Create new dataframe in which each row corresponds to failure type
# with its frequency, time_driven and distances 
failure_df = pd.DataFrame(columns=['type', 'count' , 'time(minutes)','distance(km)','intervals'])  

# loop over dictionary list values of timestamps for all failures types.
for failure, timestamps in failure_dict.items():
    # Select failed timestamps
    intervals = get_failure_intervals(df_pivot, failure, timestamps)
    # delete un-nescessary columns
    df_pivot = df_pivot.drop([failure+'_failure','timestamp','next'],axis=1)
    
    total_failed_time = 0
    for start, end in intervals:
        selected_timestamps = df_pivot.loc[[start,end]]
        _, time_in_hr = calc_time(selected_timestamps)
        total_failed_time += time_in_hr
    
    ## Append row to pie_df
    dict = {'type': failure,
            'count': len(timestamps),
            'time(minutes)': total_failed_time*60,
            'distance(km)': calc_distance(selected_timestamps,total_failed_time),
            'intervals': intervals}
    failure_df = failure_df.append(dict, ignore_index = True)

failure_df
```

```python
fig, ax = plt.subplots(1,3,figsize=(20,6))
axs = ax.ravel()
my_explode = (0, 0.1, 0,0)

ax[0].pie(failure_df['count'], 
        labels=failure_df['type'], 
        autopct='%1.2f%%',startangle=90, shadow=True, explode=my_explode)
ax[0].set_title(f'Percentage of occurence',fontsize=15)
ax[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

failure_df['time(minutes)'].plot(kind='bar', 
                                 ax=ax[1]).set_title(f"Time driven",fontsize=15)
ax[1].set_xticklabels(failure_df.type, rotation=90)


failure_df['distance(km)'].plot(kind='bar', 
                                 ax=ax[2]).set_title(f"Distances driven",fontsize=15)
ax[2].set_xticklabels(failure_df.type, rotation=90)

fig.show()
```

- Consider this problem as **Anomaly Detection**. It is the process of identifying unexpected items or events in data sets, which differ from the norm. 
- **Goal:**
    - The goal of the anomaly detection algorithm through the sensor data is to learn the patterns of a successful sensor reading so that when an anomalous activity (failure) occurs, we can flag it through the inclusion-exclusion principle
    
- **Assumptions:**
    - Each feature should be normally distributed in order to apply the unsupervised anomaly detection algorithm so features differ from the normal instances significantly.
    - Anomalies only occur very rarely in the data.

Re-use detect failure [mechanism](#mechanism) in order to label data to evaluate our model classification

```python
# Create a column called class.
# 0 (Normal): correct normal reading. 
# 1 (Anomaly): failure system reading.
classified_data = df_pivot.copy()
classified_data['class'] = 0

# Call failure mechanism to anotate anomalies readings.
failure_dict = detect_failure(classified_data)

# loop over dictionary list values of timestamps for all failures types.
for failure, timestamps in failure_dict.items():
    # set anomaly class to True for failed timestamps
    classified_data.loc[timestamps,'class']=1
```
